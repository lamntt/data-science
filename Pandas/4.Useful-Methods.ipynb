```python
def last_four(num):
    return str(num)[-4:]
```

```python
df['CC Number'].apply(last_four)
```

```python
def dollar_sign(tip):
    if tip <20:
        return '$'
    elif 20<=tip <=30:
        return '$$'
    else:
        return '$$$'
```

```python
df['Expensive'] = df['total_bill'].apply(dollar_sign)
```

```python
df
```

```python
df['total_bill'].apply(lambda bill:bill*2)
```

```python
def quality(total_bill,tip):
    if tip/total_bill>0.15:
        return 'Generous'
    else:
        return 'Other'
```

```python
df['Tip Quality'] = df[['total_bill','tip']].apply(lambda df: quality(df['total_bill'],df['tip']),axis=1)
```

```python
df['Tips Quality'] = np.vectorize(quality)(df['total_bill'],df['tip'])
```

```python
df = df.drop('Tip Quality',axis=1)
```
